## **Importing all the necessary libraries**

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
%matplotlib inline
from sklearn.preprocessing import StandardScaler
from sklearn.dummy import DummyRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import QuantileTransformer
from sklearn.model_selection import KFold
```

## **Reading and Checking our dataset**

We first read the dataset, and assigned it to the variable "df". Then we dropped any columns that were either full of missing values (such as "Population (historical)" or irrelevant ones (Such as "Code" or "Entity"))


```python
#reading the dataset and assigning it to a variable 
df = pd.read_csv("air-pollution-vs-gdp-per-capita.csv")

#Dropping useless or irrelevant columns for a cleaner dataset
df = df.drop(['Population (historical)', 'Code', 'time', 'time.1', 'time.2', 'Entity'], axis=1)

#Checking that everything was dropped correctly 
df.head()
```

```python
#Making sure that no null values were left
df.isnull().sum()
```

# **Visualising Rlationships between Columns**

- **SPM over time:** First we visualised how the overall quality of air changed in london over time
- **GDP over time:** Then we visualised the way the GDP of the country changed over the years
- **GDP + SPM over time:** We also combined the two plots, so that we can have a much more accurate and easy outlook on the contents of our data
- **GDP vs SPM scatterplot with a regplot over it:** Finally, we plotted the relatioship of GDP with SPM, and fit a cubic curve over it in order to see some patterns. Patterns that we saw: When the GDP only starts to rise, the SPM peaks, and then as the GDP continue to grow faster and faster, SPM decreases to a much lower count than at first. The plot follows the patterns of EKC (Environmental Kuznets Curve), with first a spike in the early stages, and then a slow decrease later on. There is a slight spike up in the end, but due to that not being visible on the plot we presume that it is from some kind of noise.

```python
#assigning our predictors and response variables as X and y
X = df.drop(columns=['Suspended Particulate Matter (SPM) (Fouquet and DPCC (2011))'])
y = df['Suspended Particulate Matter (SPM) (Fouquet and DPCC (2011))']
```

```python
#performing train test split to separate the data into training and testing sections
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=33)
```

```python
#initiating model 
lreg = LinearRegression()

#fitting the model over the train data 
lreg.fit(X_train, y_train)

#predicting test values with the model 
y_pred = lreg.predict(X_test)

#calculating test results 
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

#printing the test results to see what we're working with 
print(f"Linear Regression Mean Squared Error: {mse}")
print(f"Linear Regression R2 Score: {r2}")
print(f"Linear Regression Mean Absolute Error: {mae}")
print(f"Linear Regression Root Mean Square Error: {rmse}")

#initiating a kfold cross validator with 5 folds
cv = KFold(n_splits=5, shuffle=True, random_state=42)

#performing cross validation with neg_mean_squared_error and print the scores 
scores_lreg = cross_val_score(lreg, X, y, cv=cv, scoring='neg_mean_squared_error')
print("Shuffled CV scores (Linear Regression):", scores_lreg)
```

```python
#initiating model
knn = KNeighborsRegressor(n_neighbors=3)

#fitting the model over the train data 
knn.fit(X_train, y_train)

#predicting test values with the model 
y_pred = knn.predict(X_test)

#calculating test results
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

#printing the test results to see what we're working with 
print(f"knn regression model Mean Squared Error: {mse}")
print(f"knn regression model R2 Score: {r2}")
print(f"knn regression model Mean Absolute Error: {mae}")
print(f"knn regression model Root Mean Square Error: {rmse}")

#initiating a kfold cross validator with 5 folds
cv = KFold(n_splits=5, shuffle=True, random_state=42)

#performing cross validation with neg_mean_squared_error and print the scores
scores_knn = cross_val_score(knn, X, y, cv=cv, scoring='neg_mean_squared_error')
print("Shuffled CV scores (knn):", scores_knn)
```

```python
#initiating model
rf = RandomForestRegressor(n_estimators=100, random_state=42)

#fitting the model over the train data 
rf.fit(X_train, y_train)

#predicting test values with the model 
y_pred = rf.predict(X_test)

#calculating test results 
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

#printing the test results to see what we're working with
print(f"Random Forest Regression Mean Squared Error: {mse}")
print(f"Random Forest Regression R2 Score: {r2}")
print(f"Random Forest Regression Mean Absolute Error: {mae}")
print(f"Random Forest Regression Root Mean Square Error: {rmse}")

#initiating a kfold cross validator with 5 folds
cv = KFold(n_splits=5, shuffle=True, random_state=42)

#performing cross validation with neg_mean_squared_error and print the scores
scores_rf = cross_val_score(rf, X, y, cv=cv, scoring='neg_mean_squared_error')
print("Shuffled CV scores (Random Forest):", scores_rf)
```

```python
#initiating model
dtree = DecisionTreeRegressor(random_state=42)

#fitting the model over the train data 
dtree.fit(X_train, y_train)

#predicting test values with the model 
y_pred = dtree.predict(X_test)

#calculating test results 
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

#printing the test results to see what we're working with
print(f"Decision Tree Regression Mean Squared Error: {mse}")
print(f"Decision Tree Regression R2 Score: {r2}")
print(f"Decision Tree Regression Mean Absolute Error: {mae}")
print(f"Decision Tree Regression Root Mean Square Error: {rmse}")

#initiating a kfold cross validator with 5 folds
cv = KFold(n_splits=5, shuffle=True, random_state=42)

#performing cross validation with neg_mean_squared_error and print the scores
scores_dtree = cross_val_score(dtree, X, y, cv=cv, scoring='neg_mean_squared_error')
print("Shuffled CV scores (Decision Tree):", scores_dtree)
```

```python
#initiating model
br = BaggingRegressor(DecisionTreeRegressor(), n_estimators=10, random_state=42)

#fitting the model over the train data 
br.fit(X_train, y_train)

#predicting test values with the model 
y_pred = br.predict(X_test)

#calculating test results 
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

#printing the test results to see what we're working with
print(f"Bagging Model Regression Mean Squared Error: {mse}")
print(f"Bagging Model Regression R2 Score: {r2}")
print(f"Bagging Model Regression Mean Absolute Error: {mae}")
print(f"Bagging Model Regression Root Mean Square Error: {rmse}")

#initiating a kfold cross validator with 5 folds
cv = KFold(n_splits=5, shuffle=True, random_state=42)

#performing cross validation with neg_mean_squared_error and print the scores
scores_br = cross_val_score(br, X, y, cv=cv, scoring='neg_mean_squared_error')
print("Shuffled CV scores (Bagging Model):", scores_br)
```
